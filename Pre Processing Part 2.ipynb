```python
# Load the dataset
file_path = "DisneylandReviews.csv"  # Update with the correct path if needed
df = pd.read_csv(file_path, encoding="latin1")

# Remove rows where 'Year_Month' is 'missing'
df = df[df['Year_Month'] != 'missing']

# Convert 'Year_Month' to datetime format
df['Year_Month'] = pd.to_datetime(df['Year_Month'], format='%Y-%m')

# Remove duplicate rows
df = df.drop_duplicates()

# Clean 'Review_Text' (lowercase and remove special characters)
df['Review_Text'] = df['Review_Text'].str.lower().str.replace(r'[^a-z0-9\s]', '', regex=True)

# Display dataset info and sample
print(df.info())
print(df.head())
```

```python
# Convert review text to lowercase for uniform processing
df['Review_Text'] = df['Review_Text'].astype(str).str.lower()

# Define keyword-based classification
def visitor_type(review):
    if re.search(r"\bfirst time\b", review):
        return "First-Time Visitor"
    elif re.search(r"\b(again|second visit|returning|back)\b", review):
        return "Returning Visitor"
    else:
        return "Unknown"

# Apply function to create new column
df['Visitor_Type'] = df['Review_Text'].apply(visitor_type)

# Export the processed data
export_path = "C:\\Users\\Shreyas MN\\Desktop\\Uh Academics\\Information Visualization\\PROJECT 2.0\\DisneylandReviews_Processed.csv"  # Modify for local system
df.to_csv(export_path, index=False)

print("Processing Complete! Cleaned data saved as:", export_path)
```

```python
import pandas as pd
import numpy as np
import re
#import TextBlob

# Load Dataset
file_path = "DisneylandReviews.csv"  # Update with the correct path if needed
df = pd.read_csv(file_path, encoding="latin1")

#1. Date & Time Features
df['Year_Month'] = pd.to_datetime(df['Year_Month'], format='%Y-%m', errors='coerce')
df['year'] = df['Year_Month'].dt.year
df['month'] = df['Year_Month'].dt.month
df['month_name'] = df['Year_Month'].dt.month_name()
df['day_of_week'] = np.random.choice(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], size=len(df))

#2. Review Text Features
df['review_length'] = df['Review_Text'].apply(lambda x: len(str(x)))  # Length of review
df['review_word_count'] = df['Review_Text'].apply(lambda x: len(str(x).split()))  # Number of words
df['review_sentence_count'] = df['Review_Text'].apply(lambda x: len(str(x).split('.')))  # Number of sentences
df['word_density'] = df['review_word_count'] / df['review_length']  # Words per character
df['uppercase_count'] = df['Review_Text'].apply(lambda x: sum(1 for c in str(x) if c.isupper()))  # Uppercase letters
df['digit_count'] = df['Review_Text'].apply(lambda x: sum(c.isdigit() for c in str(x)))  # Number of digits
df['punctuation_count'] = df['Review_Text'].apply(lambda x: sum(1 for c in str(x) if c in "!?,."))  # Punctuation count

# 3. Sentiment & Emotion Analysis
def get_sentiment(text):
    try:
        sentiment = TextBlob(str(text)).sentiment
        return sentiment.polarity, sentiment.subjectivity
    except:
        return 0, 0

df[['sentiment_polarity', 'sentiment_subjectivity']] = df['Review_Text'].apply(lambda x: pd.Series(get_sentiment(x)))

df['review_has_exclamation'] = df['Review_Text'].apply(lambda x: 1 if '!' in str(x) else 0)
df['review_has_question'] = df['Review_Text'].apply(lambda x: 1 if '?' in str(x) else 0)
df['review_has_numbers'] = df['Review_Text'].apply(lambda x: 1 if any(c.isdigit() for c in str(x)) else 0)

#4. User & Location Features
df['location_encoded'] = df['Reviewer_Location'].astype('category').cat.codes  # Convert locations into numbers
df['location_length'] = df['Reviewer_Location'].apply(lambda x: len(str(x)))  # Length of location name

#5. Branch & Disneyland-Specific Features
df['branch_encoded'] = df['Branch'].astype('category').cat.codes  # Convert branches into numbers

#6. Rating & Engagement Features
df['rating_category'] = df['Rating'].apply(lambda x: 'Low' if x <= 2 else ('Medium' if x <= 4 else 'High'))
df['is_high_rating'] = df['Rating'].apply(lambda x: 1 if x == 5 else 0)  # 1 if the rating is 5, else 0

#7. Most Common Words in Reviews
def most_common_words(text, n=5):
    words = re.findall(r'\b\w+\b', str(text).lower())  # Extract words
    return [word for word, count in Counter(words).most_common(n)]

df['top_5_words'] = df['Review_Text'].apply(lambda x: most_common_words(x, 5))

#8. Interaction-Based Features
df['review_has_adjectives'] = df['Review_Text'].apply(lambda x: any(word in str(x).lower() for word in ['amazing', 'great', 'awesome', 'bad', 'terrible']))
df['review_has_comparison_words'] = df['Review_Text'].apply(lambda x: any(word in str(x).lower() for word in ['better', 'worse', 'best', 'worst', 'than']))

# Display extracted features
print(df.head())
```

```python

import re

def preprocess_disneyland_reviews(file_path):
    # Load dataset
    df = pd.read_csv(file_path)
    
    # 1. Handling Missing Values
    df['Year_Month'].fillna('Unknown', inplace=True)
    df['year'].fillna(df['year'].median(), inplace=True)
    df['month'].fillna(df['month'].median(), inplace=True)
    df['month_name'].fillna('Unknown', inplace=True)
    
    # Convert year and month to integer
    df['year'] = df['year'].astype(int)
    df['month'] = df['month'].astype(int)
    
    # 2. Cleaning Review Text
    def clean_text(text):
        text = str(text).lower()  # Convert to lowercase
        text = re.sub(r'[^a-z0-9.,!?\s]', '', text)  # Remove special characters
        text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces
        return text
    
    df['Review_Text'] = df['Review_Text'].apply(clean_text)
    
    # 3. Handling Outliers
    for col in ['review_length', 'review_word_count', 'review_sentence_count']:
        upper_limit = df[col].quantile(0.99)
        lower_limit = df[col].quantile(0.01)
        df[col] = np.clip(df[col], lower_limit, upper_limit)
    
    # 4. Correct Sentiment Scores if needed
    df['sentiment_polarity'] = df['sentiment_polarity'].astype(float)
    df['sentiment_subjectivity'] = df['sentiment_subjectivity'].astype(float)
    
    # Save the cleaned dataset
    cleaned_file_path = file_path.replace('.csv', '_cleaned.csv')
    df.to_csv(cleaned_file_path, index=False)
    
    return cleaned_file_path

# Update the file path with your local directory
file_path = r"C:\Users\Shreyas MN\Desktop\Uh Academics\Information Visualization\PROJECT 2.0\DisneylandReviews_Extracted.csv"

# Run the preprocessing function
cleaned_file = preprocess_disneyland_reviews(file_path)

print(f'Cleaned data saved at: {cleaned_file}')
```
